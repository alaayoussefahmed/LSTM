```python
import math

X = [1, 2, 3]

Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0

Wy, by = 4, 0

h = 0.0
C = 0.0

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

results = {}

for t, x in enumerate(X, start=1):
    f = sigmoid(Wf * x + Whf * h + bf)
    i = sigmoid(Wi * x + Whi * h + bi)
    C_tilde = tanh(Wc * x + Whc * h + bc)
    C = f * C + i * C_tilde
    o = sigmoid(Wo * x + Who * h + bo)
    h = o * tanh(C)

    results[f"step_{t}"] = {
        "x": x,
        "f": round(f, 3),
        "i": round(i, 3),
        "C_tilde": round(C_tilde, 3),
        "C": round(C, 3),
        "o": round(o, 3),
        "h": round(h, 3)
    }

x4 = 4
f4 = sigmoid(Wf * x4 + Whf * h + bf)
i4 = sigmoid(Wi * x4 + Whi * h + bi)
C_tilde4 = tanh(Wc * x4 + Whc * h + bc)
C4 = f4 * C + i4 * C_tilde4
o4 = sigmoid(Wo * x4 + Who * h + bo)
h4 = o4 * tanh(C4)
y_hat = Wy * h4 + by

results["step_4_prediction"] = {
    "f4": round(f4, 3),
    "i4": round(i4, 3),
    "C_tilde4": round(C_tilde4, 3),
    "C4": round(C4, 3),
    "o4": round(o4, 3),
    "h4": round(h4, 3),
    "y_hat": round(y_hat, 3),
    "final_prediction": round(y_hat, 1)
}

results
```
