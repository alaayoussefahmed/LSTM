```python
import math
# Input sequence
X = [1, 2, 3]
Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0
# Output layer
Wy, by = 4, 0
# Initial states
h = 0.0
C = 0.0
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
def tanh(x):
    return math.tanh(x)
#  x1, x2, x3
for t, x in enumerate(X, start=1):
 f = sigmoid(Wf * x + Whf * h + bf)
 i = sigmoid(Wi * x + Whi * h + bi)
 C_tilde = tanh(Wc * x + Whc * h + bc)
 C = f * C + i * C_tilde
 o = sigmoid(Wo * x + Who * h + bo)
 h = o * tanh(C)
 print(f"Time step {t}, x = {x}")
 print(f"f{t} = {f:.3f}")
 print(f"i{t} = {i:.3f}")
 print(f"C~{t} = {C_tilde:.3f}")
 print(f"C{t} = {C:.3f}")
 print(f"o{t} = {o:.3f}")
 print(f"h{t} = {h:.3f}")
 print("" * 30)
# Predict x4
x4 = 4
f4 = sigmoid(Wf* x4 + Whf * h + bf)
i4 = sigmoid(Wi * x4 + Whi * h + bi)
C_tilde4 = tanh(Wc * x4 + Whc * h + bc)
C4 = f4 * C + i4 * C_tilde4
o4 = sigmoid(Wo * x4 + Who * h + bo)
h4 = o4 * tanh(C4)
# Final prediction
y_hat = Wy * h4 + by
print("Prediction step (t = 4)")
print(f"f4 = {f4:.3f}")
print(f"i4 = {i4:.3f}")
print(f"C~4 = {C_tilde4:.3f}")
print(f"C4 = {C4:.3f}")
print(f"o4 = {o4:.3f}")
print(f"h4 = {h4:.3f}")
print(f"y_hat = {y_hat:.3f}")
print(f"Final Prediction ≈ {round(y_hat, 1)}")
```
